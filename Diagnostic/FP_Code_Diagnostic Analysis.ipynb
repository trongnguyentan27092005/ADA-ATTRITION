### 4.1.4 Problem 4 (Age, Education)

Given that attrition is concentrated among younger employees, does educational attainment interact with age to further explain variations in attrition risk?

```python
# Filter the dataset to focus on younger employees (aged 25–35), the age group where attrition has been identified as most concentrated
young_df = df[(df['Age'] >= 25) & (df['Age'] <= 35)]

# Create a normalized cross-tabulation to examine how attrition rates vary across different levels of educational attainment within younger employees
pd.crosstab(
    young_df['Education'],
    young_df['Attrition'],
    normalize='index'
)

# Visualize attrition and non-attrition counts by education level to identify patterns of attrition risk among younger employees
sns.countplot(
    data=young_df,
    x='Education',
    hue='Attrition'
)
plt.title('Attrition by Education Level among Younger Employees (Age 25–35)')
plt.xticks(rotation=30)
plt.show()

# Perform a Chi-square test of independence to assess whether educational attainment and attrition are statistically associated within the younger employee group
from scipy.stats import chi2_contingency

table = pd.crosstab(
    young_df['Education'],
    young_df['Attrition']
)

chi2_contingency(table)
```

Among younger employees (25–35):

*   The number of job resignations was higher among Bachelor and Master graduates. However, these are also the two largest groups.
*   The Below College and Doctor groups had very few job turnovers → small sample size
--> Visually, differences in quantity are not sufficient to conclude differences in the risk of employee turnover.




p_value=0.77>0.05 -> These differences are not statistically significant. Therefore, educational attainment does not significantly interact with age to explain attrition risk among younger employees.

### 4.2.3 Problem 3 (DistanceFromHome, JobRole)

Is there a statistically significant difference in monthly income between the group that quits and the group that stays when analyzed by JobRole or JobLevel?


```python
sns.boxplot(
    data=df,
    x='JobLevel',
    y='MonthlyIncome',
    hue='Attrition'
)
plt.title('Monthly Income by Attrition across Job Levels')
plt.show()
```



*   Within the same JobLevel, the group who quit (Attrition = Yes) and those who stayed (No) had fairly similar income levels, with no large or clear gap.
*   Income increased primarily based on JobLevel, not on the number of people who quit.

*  Attrition was not strongly concentrated in the low-income group when controlled by JobLevel.

```python
# Since Monthly Income is continuous, Attitude is divided into two groups -->Do the t-test in each Job Level
from scipy.stats import ttest_ind

for level in sorted(df['JobLevel'].unique()):
    group = df[df['JobLevel'] == level]

    income_yes = group[group['Attrition'] == 'Yes']['MonthlyIncome']
    income_no  = group[group['Attrition'] == 'No']['MonthlyIncome']

    if len(income_yes) > 10 and len(income_no) > 10:
        stat, p = ttest_ind(income_yes, income_no, equal_var=False)
        print(f'JobLevel {level}: p-value = {p:.4f}')
```

Income-related attrition is significantly significant only at the Entry Level, while no significant income difference is observed at higher job levels.

```python
high_attr_roles = (
    df[df['Attrition'] == 'Yes']['JobRole']
    .value_counts()
    .head(3)
    .index
)

for role in high_attr_roles:
    group = df[df['JobRole'] == role]

    income_yes = group[group['Attrition'] == 'Yes']['MonthlyIncome']
    income_no  = group[group['Attrition'] == 'No']['MonthlyIncome']

    if len(income_yes) > 10 and len(income_no) > 10:
        stat, p = ttest_ind(income_yes, income_no, equal_var=False)
        print(f'{role}: p-value = {p:.4f}')
```

Income-related attrition is statistically significant for technical roles such as Laboratory Technician and Research Scientist, but not for Sales Executive, indicating that the impact of income on attrition varies by JobRole.

The impact of monthly income on attrition is not uniform. It is statistically significant for entry-level employees and certain technical roles, but not for higher job levels or sales roles, suggesting heterogeneity in attrition drivers across employee segments.

*   The distribution of those leaving the company shifts to the right → living further away from the company.

Human Resources


*   The gap between those who quit and those who stayed is larger and more dispersed than the gap between those who remained.



JobRole another (Research Scientist, Lab Technician, Sales Rep)


*   The two Yes/No groups overlap strongly.
*   Medians are close together → it is difficult to see the difference visually.




```python
# Test thống kê theo từng JobRole (Vì DistanceFromHome thường không chuẩn, dùng Mann–Whitney U test)
from scipy.stats import mannwhitneyu

results = []

for role in df['JobRole'].unique():
    quit_group = df[(df['JobRole'] == role) & (df['Attrition'] == 'Yes')]['DistanceFromHome']
    stay_group = df[(df['JobRole'] == role) & (df['Attrition'] == 'No')]['DistanceFromHome']

    if len(quit_group) > 10 and len(stay_group) > 10:
        stat, p = mannwhitneyu(quit_group, stay_group, alternative='two-sided')
        results.append([role, p])

pd.DataFrame(results, columns=['JobRole', 'p_value']).sort_values('p_value')
```

Sales Executive (p = 0.0036) and Human Resources (p = 0.0143)

→ Distance from home to work is significantly related to attrition.
The group that quit (Attrition = Yes) in these roles tended to live further from the company than the group that stayed.




Sales Representative, Research Scientist, Laboratory Technician (p-value > 0.05)

→ There is no statistical evidence to suggest that distance affects the decision to quit.
The home-to-work distance is not a clear differentiating factor between the quitting and staying groups in these roles.

At low Job Involvement:

- Low/Medium satisfaction → significantly higher turnover rate.

At high/very high Job Involvement:
- Attrition is always low, even when Job Satisfaction is not high.
- Job Involvement acts as a buffer (protective factor).

Attrition is not driven by job dissatisfaction alone. Employees who are both disengaged (low job involvement) and dissatisfied exhibit the highest attrition risk, while high job involvement significantly mitigates turnover even under lower satisfaction levels.
