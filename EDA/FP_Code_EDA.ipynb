```python
# Function to analyse the missing data

def analyze_null_values(df):
    missing_data = df.isnull().sum().to_frame(name="Null Values")
    missing_data["Total Values"] = len(df)
    missing_data["Percentage of Missing Values"] = (missing_data["Null Values"] / missing_data["Total Values"]) * 100
    return missing_data.sort_values(by="Percentage of Missing Values", ascending=False)

# Example usage
analyze_null_values(df)
```

*Binning for categorical variables*

```python
# Binning for categorical variables

# Dictionaries based on Kaggle dataset description
# Education
education_map = {1:'Below College', 2:'College', 3:'Bachelor', 4:'Master', 5:'Doctor'}

# EnvironmentSatisfaction, JobInvolvement, JobSatisfaction, RelationshipSatisfaction
satisfaction_map = {1:'Low', 2:'Medium', 3:'High', 4:'Very High'}

# PerformanceRating
performance_map = {1:'Low', 2:'Good', 3:'Excellent', 4:'Outstanding'}

# WorkLifeBalance
worklife_map = {1:'Bad', 2:'Good', 3:'Better', 4:'Best'}

# JobLevel - Based on https://www.earnin.com/blog/job-levels
joblevel_map = {1:'Entry Level', 2:'Mid Level', 3:'Senior', 4:'Director', 5:'Executive'}
df['JobLevel'] = df['JobLevel'].map(joblevel_map).astype('category')

# StockOptionLevel - No references, just conclude through information found in internet
stockoption_map = {0:'None', 1:'Low', 2:'Medium', 3:'High'}
df['StockOptionLevel'] = df['StockOptionLevel'].map(stockoption_map).astype('category')


# Apply mappings
df['Education'] = df['Education'].map(education_map)
df['EnvironmentSatisfaction'] = df['EnvironmentSatisfaction'].map(satisfaction_map)
df['JobInvolvement'] = df['JobInvolvement'].map(satisfaction_map)
df['JobSatisfaction'] = df['JobSatisfaction'].map(satisfaction_map)
df['RelationshipSatisfaction'] = df['RelationshipSatisfaction'].map(satisfaction_map)
df['PerformanceRating'] = df['PerformanceRating'].map(performance_map)
df['WorkLifeBalance'] = df['WorkLifeBalance'].map(worklife_map)

# Convert to categorical dtype
ordinal_cols = [
    'Education','EnvironmentSatisfaction','JobInvolvement','JobLevel',
    'JobSatisfaction','PerformanceRating','RelationshipSatisfaction',
    'StockOptionLevel','WorkLifeBalance'
]
df[ordinal_cols] = df[ordinal_cols].astype('category')
```

## **3.3 - Univariate analysis**

### **3.3.1 - UA|Quantitative variables**

```python
# Histogram vissualization for numeric variables
numeric_cols = df_numeric.columns.tolist()

n_cols = 4
n_rows = math.ceil(len(numeric_cols) / n_cols)

fig, ax = plt.subplots(n_rows, n_cols, figsize=(25, 5 * n_rows))

ax = ax.flatten()  # trải phẳng mảng 2D thành 1D

for i, col in enumerate(numeric_cols):
    sns.histplot(df_numeric[col].dropna(), kde=True, ax=ax[i])
    ax[i].set_title(col)

# Xóa subplot trống (nếu có)
for j in range(i + 1, n_rows * n_cols):
    fig.delaxes(ax[j])

plt.tight_layout()
plt.show()
```

```python
# Box plot chart for numeric variables
plt.figure(figsize=(18, 20))

for i, col in enumerate(continuous_cols, 1):
    plt.subplot(5, 3, i)  # 5 rows x 3 columns grid
    sns.boxplot(y=df_numeric[col], color="skyblue")
    plt.title(col)

plt.tight_layout()
plt.show()
```

### **3.3.2 - UA|Quanlitative variables**

```python
import math
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")

n_cols = 4
n_rows = math.ceil(len(categorical_cols) / n_cols)

plt.figure(figsize=(26, n_rows * 5))

for i, col in enumerate(categorical_cols, 1):
    plt.subplot(n_rows, n_cols, i)

    # Tạo các mức xanh theo số lượng category trong từng cột
    n_unique = df_category[col].nunique()
    palette = sns.color_palette("Blues", n_unique)

    ax = sns.countplot(
        data=df_category,
        x=col,
        palette=palette
    )

    plt.title(col, fontsize=13, weight="bold")
    plt.xticks(rotation=35, ha="right", fontsize=9)

    # Hiện số trên đỉnh cột
    for p in ax.patches:
        height = p.get_height()
        ax.annotate(
            f'{int(height)}',
            (p.get_x() + p.get_width()/2, height),
            ha='center',
            va='bottom',
            fontsize=9,
            weight='bold'
        )

plt.tight_layout(pad=3)
plt.show()
```

 The workforce is concentrated in R&D, entry and mid-level roles, and bachelor-level education, with generally high satisfaction and involvement. However, overtime, lack of stock options, and poor work-life balance stand out as potential attrition triggers.

## **3.5 - Analyzing data align with 3 segments**

### **Segment 1: Demographics**

```python
# Segment 1: Demographics
demo_cols = ['Age','Gender','MaritalStatus','Education','EducationField']

n_cols = 2
n_rows = math.ceil(len(demo_cols) / n_cols)

fig, ax = plt.subplots(n_rows, n_cols, figsize=(14, 5 * n_rows))
ax = ax.flatten()

for i, col in enumerate(demo_cols):
    if df[col].dtype == 'object' or str(df[col].dtype) == 'category':
        sns.countplot(data=df, x=col, hue="Attrition", palette="Set2", ax=ax[i])
        ax[i].tick_params(axis='x', rotation=45)
    else:
        sns.boxplot(data=df, x="Attrition", y=col, palette="Set2", ax=ax[i])
    ax[i].set_title(f"{col} vs Attrition")

# Remove empty subplots
for j in range(i+1, n_rows*n_cols):
    fig.delaxes(ax[j])

plt.tight_layout()
plt.show()
```

- Age: Employees who left are generally younger than those who stayed.

- Gender: Slightly more male employees left compared to females.

- Marital Status: Single employees show higher attrition than married or divorced.

- Education: Employees with lower education levels (Below College, College) tend to leave more.

- Education Field: Employees from Marketing, Human Resources, and Other fields show lower attrition than those from Life Sciences or Medical.

--> **Younger, single, less-educated employees from non-technical fields are more likely to leave**
